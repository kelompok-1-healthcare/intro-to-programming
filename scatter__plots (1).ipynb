Scatter plot, atau diagram sebar, adalah jenis grafik yang digunakan untuk memvisualisasikan hubungan antara dua variabel numerik.

##Scatter plots

Kode ini menghasilkan scatter plot untuk memvisualisasi hubungan antara Body Mass Index (BMI) dan biaya asuransi dalam dataset insurance_data.

```python
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'])
```

* sns.scatterplot: Fungsi dari pustaka Seaborn untuk membuat scatter plot.
* x=insurance_data['bmi']: Menetapkan kolom bmi dari DataFrame insurance_data sebagai sumbu x.
* y=insurance_data['charges']: Menetapkan kolom charges dari DataFrame insurance_data sebagai sumbu y.

```python
sns.regplot(x=insurance_data['bmi'], y=insurance_data['charges'])
```



*   sns.regplot: Fungsi dari pustaka Seaborn untuk membuat regresi plot.
*   x=insurance_data['bmi']: Menetapkan kolom bmi dari DataFrame insurance_data sebagai sumbu x.
*   y=insurance_data['charges']: Menetapkan kolom charges dari DataFrame insurance_data sebagai sumbu y.

  



##Color-coded scatter plots

Kode ini menghasilkan scatter plot dengan hue (warna) untuk memvisualisasi hubungan antara Body Mass Index (BMI), biaya asuransi, dan status perokok dalam dataset insurance_data. Scatter plot dengan hue memungkinkan Anda untuk melihat bagaimana hubungan antara BMI dan biaya asuransi bervariasi berdasarkan status perokok.

```python
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'], hue=insurance_data['smoker'])
```

1. Fungsi sns.scatterplot digunakan untuk membuat scatter plot dengan hue.
2. Kolom bmi dari DataFrame insurance_data diplot pada sumbu x.
3. Kolom charges dari DataFrame insurance_data diplot pada sumbu y.
4. Kolom smoker dari DataFrame insurance_data digunakan untuk mewarnai titik data.

```python
sns.lmplot(x="bmi", y="charges", hue="smoker", data=insurance_data)
```

1. sns.lmplot: Fungsi dari pustaka Seaborn untuk membuat plot regresi linier dengan hue.
2. x="bmi": Menetapkan kolom bmi dari DataFrame insurance_data sebagai sumbu x.
3. y="charges": Menetapkan kolom charges dari DataFrame insurance_data sebagai sumbu y.
4. hue="smoker": Menetapkan kolom smoker dari DataFrame insurance_data sebagai hue. Hue mewarnai setiap garis regresi dan titik data berdasarkan status perokok (perokok atau bukan perokok).
5. data=insurance_data: Menentukan DataFrame insurance_data sebagai sumber data untuk plot.

```python
sns.swarmplot(x=insurance_data['smoker'],
              y=insurance_data['charges'])
```

Kode ini menghasilkan swarmplot untuk memvisualisasi hubungan antara status perokok (smoker) dan biaya asuransi (charges) dalam dataset insurance_data. Swarmplot adalah jenis scatterplot yang memisahkan data berdasarkan kategori (dalam hal ini, perokok dan bukan perokok) dan kemudian menampilkan distribusi data dalam setiap kategori.

1. sns.swarmplot: Fungsi dari pustaka Seaborn untuk membuat swarmplot.
2. x=insurance_data['smoker']: Menetapkan kolom smoker dari DataFrame insurance_data sebagai sumbu x.
3. y=insurance_data['charges']: Menetapkan kolom charges dari DataFrame insurance_data sebagai sumbu y.
